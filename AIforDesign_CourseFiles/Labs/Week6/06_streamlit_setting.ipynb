```python
%%writefile media_demo.py
import streamlit as st

# Set the page configuration
st.set_page_config(page_title="Multimedia Demo", layout="wide")

st.title("🎨 Part 2: Exploring Images, Video & Audio")
st.write("Let's experiment with different ways to load and display media files in Streamlit.")

st.divider()

# --- 1. Basic Image Loading & Sizing ---
st.header("1. Image Loading & Sizing")
st.write("When showcasing your AI-generated images, controlling the display size is crucial.")

st.subheader("Responsive Size")
# use_container_width makes the image adapt to the screen/container size automatically
st.image("https://picsum.photos/id/16/800/400", caption="Adapts to screen width (use_container_width=True)", use_container_width=True)

st.subheader("Fixed Size")
# Force the image to be exactly 300 pixels wide
st.image("https://picsum.photos/id/16/800/400", caption="Fixed at 300 pixels wide", width=300)

st.divider()

# --- 2. Hiding Media in Expanders ---
st.header("2. Keeping UI Clean with Expanders")
st.write("If you have heavy video files or a long list of generated assets, hide them inside an expander!")

# Use an expander to hide the video until clicked
with st.expander("🎬 Click here to reveal the hidden video"):
    st.video("https://www.w3schools.com/html/mov_bbb.mp4")
    st.caption("Expanders are perfect for optional content or saving screen space.")

st.divider()

# --- 3. Audio Support ---
st.header("3. Adding Audio")
st.write("You can also add background music or AI-generated voiceovers to your app.")

# Load a sample audio file
st.audio("https://www.w3schools.com/html/horse.ogg", format="audio/ogg")

st.info("💡 Tip: Try replacing the URLs above with the links to your own images or videos!")
```

```python
%%writefile interactive_demo.py
import streamlit as st
import time

# Set the page configuration
st.set_page_config(page_title="Interactive Demo", layout="centered")

st.title("🕹️ Part 3: Interactive Logic")
st.write("Let's make our app dynamic! Here are 4 practical examples of user interaction.")

st.divider()

# --- Demo 1: The Magic Button ---
st.header("1. Buttons & Actions")
st.write("Buttons are the simplest way to trigger an event.")

# if a button is clicked, it returns True
if st.button("🚀 Click me for a surprise!"):
    st.balloons() # This triggers a fun balloon animation
    st.success("Yay! You just triggered an action.")

st.divider()

# --- Demo 2: Text Input (Simulating AI Prompt) ---
st.header("2. Text Input")
st.write("Perfect for capturing user prompts for AI models.")

# Get user input with a placeholder
user_prompt = st.text_input("Enter your image generation prompt:", placeholder="e.g., A futuristic city at sunset...")

# Only run this block if the user has typed something
if user_prompt:
    # Use a spinner to simulate loading/generation time
    with st.spinner('Simulating AI generation...'):
        time.sleep(1.5) # Simulate processing time
        st.write(f"**Your Prompt:** {user_prompt}")
        st.info("Here is where your AI-generated image would appear!")

st.divider()

# --- Demo 3: Dropdowns & Sliders (Simulating Model Parameters) ---
st.header("3. Controls & Parameters")
st.write("Use these to control model parameters or filter content in your app.")

# Selectbox for category or style
art_style = st.selectbox(
    "Choose an Art Style:",
    ["Cyberpunk", "Watercolor", "Photorealistic", "Anime"]
)

# Slider for numerical values (e.g., simulating a CFG scale in diffusion models)
cfg_scale = st.slider("Creativity Level (CFG Scale):", min_value=1, max_value=20, value=7)

# Display the real-time changes
st.write(f"**Current Settings:** Generating a **{art_style}** image with a creativity level of **{cfg_scale}**.")

st.divider()

# --- Demo 4: Memory / State Management ---
st.header("4. Memory (Session State)")
st.write("Streamlit reruns the script on every interaction. We use `st.session_state` to remember things (like a 'Like' counter).")

# Initialize the counter in session state if it doesn't exist yet
if 'like_count' not in st.session_state:
    st.session_state.like_count = 0

# Create a layout with an image and a like button next to it
col1, col2 = st.columns([1, 2])

with col1:
    st.image("https://picsum.photos/id/1025/300/300", use_container_width=True)

with col2:
    st.subheader("Cute Pug Photo")
    st.write("Do you like this image?")

    # The button displays the current like count
    if st.button(f"❤️ Like ({st.session_state.like_count})"):
        st.session_state.like_count += 1
        st.rerun() # Refresh the page immediately to show the updated count
```

```python
%%writefile comparison_template.py
import streamlit as st
from streamlit_image_comparison import image_comparison
import os

# Set the page configuration
st.set_page_config(page_title="YOLO Detection Demo", page_icon="🔍", layout="wide")

st.title("🔍 YOLO Object Detection Comparison")
st.write("Using local image files downloaded directly via wget!")
st.divider()

# =====================================================================
# STUDENT EDIT ZONE
# Define the paths to the downloaded local images
# =====================================================================
IMAGE_1 = "/content/original.jpg"
IMAGE_2 = "/content/yolo.jpg"

# Check if the images exist to prevent crashes
if not os.path.exists(IMAGE_1) or not os.path.exists(IMAGE_2):
    st.error("🚨 Image files not found! Please run the wget commands to download them first.")
    st.stop()

LABEL_1 = "Original Image"
LABEL_2 = "YOLO Detections"
# =====================================================================

st.subheader(f"Model Inference: {LABEL_1} vs {LABEL_2}")

# Use the image comparison component directly with LOCAL paths
image_comparison(
    img1=IMAGE_1,
    img2=IMAGE_2,
    label1=LABEL_1,
    label2=LABEL_2,
    width=800,
    starting_position=50,
    show_labels=True,
    make_responsive=True,
    in_memory=False # Core parameter: Must be False when using local paths
)

st.divider()
st.info("💡 Tip: Drag the slider to see exactly how the YOLO model identifies the objects!")
```
